```python
embedding_df = embedding_df.reset_index(drop=True) # 重設index
```

```python
embedding_df.info()
```

```python
#因為季別欄位只有一個是空值 所以直接人工去看他的季節是什麼填上去
embedding_df['季別'].loc[embedding_df['季別'].isna()] = "冬"
```

將所有'/,(,),|,nan'去除

```python
label_type_list = embedding_df['標籤類型'].tolist()
dimension_list = embedding_df['Dimension'].tolist()
element_list = embedding_df['元素'].tolist()
style_list = embedding_df['風格'].tolist()
material_list = embedding_df['Material'].tolist()
pattern_list = embedding_df['花型'].tolist()

for i in range(7073):
    label_type_list[i] = str(label_type_list[i]).replace('|', '').replace('nan','').replace('/','')
    dimension_list[i] = str(dimension_list[i]).replace('nan','')
    element_list[i] = str(element_list[i]).replace('|','').replace('nan','').replace('(','').replace(')','')
    style_list[i] = str(style_list[i]).replace('|', '').replace('nan','')
    material_list[i] = str(material_list[i]).replace('|', '').replace('/','').replace('(','').replace(')','').replace(',','').replace('nan','')
    pattern_list[i] = str(pattern_list[i]).replace('|','').replace('/','').replace('nan','')
```

```python
embedding_df['標籤類型'] = label_type_list
embedding_df['Dimension'] = dimension_list
embedding_df['元素'] = element_list
embedding_df['風格'] = style_list
embedding_df['Material'] = material_list
embedding_df['花型'] = pattern_list
```
